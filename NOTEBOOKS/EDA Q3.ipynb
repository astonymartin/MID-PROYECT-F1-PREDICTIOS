```python
df1.head(60).sort_values(by=["year", "q1"], ascending=[True, True])
```

```python
df1.rename(columns={'q3_x': 'q3'}, inplace=True)
df1 = df1.drop("q3_y", axis=1)
```

```python
df1['q3disqualified'] = (df1['position_q2'] > 10) * 1
df1
```

```python
df1["driver_q3vsq2"] = df1["q3"] - df1["q2"]
df1.loc[df1["q3"] == 0, "driver_q3vsq2"] = 0
df1
```

```python
car_q3vsq2 = df1.groupby(['constructorid', "year", 'circuitref'])['driver_q3vsq2'].mean()
car_q3vsq2 = car_q3vsq2.reset_index()
df1 = pd.merge(df1, car_q3vsq2, on=['constructorid', "year", 'circuitref'])
df1
```

```python
df1 = df1.rename(columns={'driver_q3vsq2_x': 'driver_q3vsq2', 'driver_q3vsq2_y': 'car_q3vsq2'})
```

```python
df1['previous_driver_q3vsq2'] = df1.groupby(['circuitref', 'driverid'])['driver_q3vsq2'].shift()
df1.previous_driver_q3vsq2 = df1.previous_driver_q3vsq2.fillna(0)
df1
```

```python
max_q3 = df1['q3'].max()
df1['q3'].replace(0, max_q3, inplace=True)
```

```python
df1['previous_car_q3vsq2'] = df1.groupby(['circuitref', 'driverid'])['car_q3vsq2'].shift()
df1.previous_car_q3vsq2 = df1.previous_car_q3vsq2.fillna(0)
df1
```

```python
df1['position_q3'] = df1.groupby(['circuitref', 'year'])['q3'].rank()
df1
```

```python
df1['previous_year_position_q3'] = df1.groupby(['circuitref', 'driverid'])['position_q3'].shift()
df1.previous_year_position_q3 = df1.previous_year_position_q3.fillna(20)
df1
```

```python
    X = dfq3.drop(drop_list, axis=1)
    y = dfq3["q3"]
    X_2024 = X[dfq3['year'] == 2024]
    y_2024 = y[dfq3['year'] == 2024]
    X_train, X_test, y_train, y_test = train_test_split(X[dfq3['year'] != 2024], 
                                                        y[dfq3['year'] != 2024], 
                                                        test_size=0.02, 
                                                        random_state=0)
    print(f'100% of our data: {len(dfq3)}.')
    print(f'80% for training data: {len(X_train)}.')
    print(f'20% for test data: {len(X_test)}.')
    
    model = LinearRegression()
    model.fit(X_train, y_train)
    predictions = model.predict(X_test)
    predictions_linear_2024  = model.predict(X_2024)
    series_pred_2024_1 = pd.Series(predictions_linear_2024, name='predictions_real_linearregresion')
    
    r2_3 = r2_score(y_test, predictions)
    RMSE_3 = mean_squared_error(y_test, predictions, squared=False)
    MSE_3 = mean_squared_error(y_test, predictions)
    MAE_3 = mean_absolute_error(y_test, predictions)
    
    ridge = Ridge()
    ridge.fit(X_train, y_train)
    predictions_ridge = ridge.predict(X_test)
    predictions_ridge_2024 = ridge.predict(X_2024)
    series_pred_2024_2 = pd.Series(predictions_ridge_2024, name='predictions_real_ridge')
  
    r2_3_2 = r2_score(y_test, predictions_ridge)
    RMSE_3_2 = mean_squared_error(y_test, predictions_ridge, squared=False)
    MSE_3_2 = mean_squared_error(y_test, predictions_ridge)
    MAE_3_2 = mean_absolute_error(y_test, predictions_ridge)
    
    lasso = Lasso()
    lasso.fit(X_train, y_train)
    predictions_lasso = lasso.predict(X_test)
    predictions_lasso_2024 = lasso.predict(X_2024)
    series_pred_2024_3 = pd.Series(predictions_lasso_2024, name='predictions_real_lasso')

    r2_3_3 = r2_score(y_test, predictions_lasso)
    RMSE_3_3 = mean_squared_error(y_test, predictions_lasso, squared=False)
    MSE_3_3 = mean_squared_error(y_test, predictions_lasso)
    MAE_3_3 = mean_absolute_error(y_test, predictions_lasso)
    
    tree = DecisionTreeRegressor()
    tree.fit(X_train, y_train)
    predictions_tree = tree.predict(X_test)
    predictions_tree_2024 = tree.predict(X_2024)
    series_pred_2024_4 = pd.Series(predictions_tree_2024, name='predictions_real_tree')

    r2_3_4 = r2_score(y_test, predictions_tree)
    RMSE_3_4 = mean_squared_error(y_test, predictions_tree, squared=False)
    MSE_3_4 = mean_squared_error(y_test, predictions_tree)
    MAE_3_4 = mean_absolute_error(y_test, predictions_tree)
    
    knn = KNeighborsRegressor()
    knn.fit(X_train, y_train)
    predictions_knn = knn.predict(X_test)
    predictions_knn_2024 = knn.predict(X_2024)
    series_pred_2024_5 = pd.Series(predictions_knn_2024, name='predictions_real_knn')

    r2_3_5 = r2_score(y_test, predictions_knn)
    RMSE_3_5 = mean_squared_error(y_test, predictions_knn, squared=False)
    MSE_3_5 = mean_squared_error(y_test, predictions_knn)
    MAE_3_5 = mean_absolute_error(y_test, predictions_knn)
    
    xgbr = xgb.XGBRFRegressor()
    xgbr.fit(X_train, y_train)
    predictions_xgb = xgbr.predict(X_test)
    predictions_xgb_2024  = xgbr.predict(X_2024)

    r2_3_6 = r2_score(y_test, predictions_xgb)
    RMSE_3_6 = mean_squared_error(y_test, predictions_xgb, squared=False)
    MSE_3_6 = mean_squared_error(y_test, predictions_xgb)
    MAE_3_6 = mean_absolute_error(y_test, predictions_xgb)

    series_pred = pd.Series(predictions_xgb, name='predictions_test')
    series_y_test = pd.Series(y_test, name='overall')
    series_pred_2024 = pd.Series(predictions_xgb_2024, name='predictions_real_xgb')
    comparison_df = pd.concat([series_y_test.reset_index(drop=True), series_pred, series_pred_2024, 
                               series_pred_2024_1, series_pred_2024_2, series_pred_2024_3, 
                              series_pred_2024_4, series_pred_2024_5], axis=1)
    df_bahrain_2024 = pd.concat([series_pred_2024.reset_index(drop=True), 
                               series_pred_2024_1, series_pred_2024_2, series_pred_2024_3, 
                              series_pred_2024_4, series_pred_2024_5], axis=1)
comparison_df
```
