```python
data1 = pd.read_csv('df_driverid_name.csv')
data3 = pd.read_csv('dataframe_trabajar_posiciones_finales.csv')
data4 = pd.read_csv('df1_bahrain_q2_predicted.csv')
data5 = pd.read_csv('df1_bahrain_q3_predicted.csv')
data6 = pd.read_csv("dfq3.csv")
```

```python
df1 = data1.copy()
df3 = data3.copy()
df4 = data4.copy()
df5 = data5.copy()
df6 = data6.copy()
```

```python
df7 = df6[df6["circuitref"] == "bahrain"]
df3 = df3.sort_values(by=['year', 'driverid'], ascending=[True, True])
df7 = df7.sort_values(by=['year', 'driverid'], ascending=[True, True])
df3 = pd.merge(df3, df7[['driverid', 'year', 'q3']], on=['driverid', 'year'], how='left')
```

```python
df4 = df4[df4["year"] == 2024]
df5 = df5[df5["year"] == 2024]
```

```python
# Crear un diccionario vac√≠o
driver_q2 = {}

# Iterar sobre cada fila del DataFrame y agregar al diccionario
for index, row in df4.iterrows():
    driverid = row['driverid']
    q2 = row['q2']
    driver_q2[driverid] = q2

# Mostrar el diccionario resultante
driver_q2

# Crear un diccionario vac√≠o
driver_q3 = {}

# Iterar sobre cada fila del DataFrame y agregar al diccionario
for index, row in df5.iterrows():
    driverid = row['driverid']
    q3 = row['q3']
    driver_q3[driverid] = q3

# Mostrar el diccionario resultante
driver_q3
```

```python
df3['q2'] = df3['q2'].fillna(df3['driverid'].map(driver_q2))
df3['q3'] = df3['q3'].fillna(df3['driverid'].map(driver_q3))
```

```python
df3['position_q2'] = df3.groupby(['circuitref', 'year'])['q2'].rank()
df3['q3disqualified'] = (df3['position_q2'] > 10) * 1
```

```python
df3['position_q3'] = df3.groupby(['circuitref', 'year'])['q3'].rank()
```

```python
df3['position_final'] = df3['position_q3']
df3['position_final'].fillna(df3['position_q2'], inplace=True)
df3['position_final'].fillna(df3['position_q1'], inplace=True)
df3['time'] = df3['q3']
df3['time'].fillna(df3['q2'], inplace=True)
df3['time'].fillna(df3['q1'], inplace=True)
```

```python
driverid_to_driverref = {}
for index, row in df1.iterrows():
    driverid = row['driverid']
    driverref = row['driverref']
    driverid_to_driverref[driverid] = driverref
df3['driver_name'] = df3['driverid'].map(driverid_to_driverref)
```

```python
circuitos = ['losail', 'bahrain', 'imola', 'catalunya', 'monaco', 'baku',
             'red_bull_ring', 'silverstone', 'hungaroring', 'spa', 'zandvoort',
             'monza', 'americas', 'rodriguez', 'interlagos',
             'yas_marina', 'albert_park', 'miami', 'villeneuve', 'marina_bay', 'suzuka']
file_names = []
nombre_base = 'df1_losail_q1_predicted_trabajar_posiciones_finales_penultimo.csv'
for circuito in circuitos:
    nombre_modificado = nombre_base.replace("losail", circuito)
    file_names.append(nombre_modificado)
dfs = {}
for file_name in file_names:
    track_name = file_name.replace(".csv", "")
    dfs[track_name] = pd.read_csv(file_name)
```

```python
file_names2 = []
nombre_base2 = 'df1_losail_q1_predicted_trabajar_posiciones_finales_q3_predicted_2024.csv'
for circuito in circuitos:
    nombre_modificado2 = nombre_base2.replace("losail", circuito)
    file_names2.append(nombre_modificado2)
dfs2 = {}
for file_name2 in file_names2:
    track_name2 = file_name2.replace(".csv", "")
    dfs2[track_name2] = pd.read_csv(file_name2)
```

```python
dfs2["df1_losail_q1_predicted_trabajar_posiciones_finales_q3_predicted_2024"]
```

```python
dfs["df1_losail_q1_predicted_trabajar_posiciones_finales_penultimo"]
```

```python
dataframes_pairs = list(zip(dfs, dfs2))
```

```python
for df1_name, df2_name in dataframes_pairs:
    # Obtener los DataFrames correspondientes
    df1 = dfs[df1_name]
    df2 = dfs2[df2_name]
    
    # Agrupar el segundo DataFrame por 'driverid' y 'year' y seleccionar las columnas relevantes
    grouped = df2.groupby(['driverid', 'year'])[['q3', 'position_q3']]
    
    # Iterar sobre los grupos
    for (driverid, year), group_df in grouped:
        # Filtrar el primer DataFrame para las filas que coincidan con el 'driverid' y 'year' actuales
        mask = (df1['driverid'] == driverid) & (df1['year'] == year)
        # Asignar los valores correspondientes del segundo DataFrame al primer DataFrame
        df1.loc[mask, 'q3'] = group_df['q3'].values
        df1.loc[mask, 'position_q3'] = group_df['position_q3'].values

        # Actualizar el DataFrame en dfs
        dfs[df1_name] = df1
```

```python
dfs["df1_losail_q1_predicted_trabajar_posiciones_finales_penultimo"]
```

```python
for track_name, df_track_name in dfs.items():
    
    df_track_name['q3'] = df_track_name['q3'].fillna(0)
    
    df_track_name['q2'] = df_track_name['q2'].replace(0, np.nan)
    df_track_name['position_q2'] = df_track_name['q2'].rank(method='min', na_option='bottom')

    
    
    # Asignar la columna 'position_q3' a 'position_final'
    df_track_name['position_final'] = df_track_name['position_q3']
    # Rellenar los valores faltantes en 'position_final' con los de 'position_q2'
    df_track_name['position_final'].fillna(df_track_name['position_q2'], inplace=True)
    # Rellenar los valores faltantes en 'position_final' con los de 'position_q1'
    df_track_name['position_final'].fillna(df_track_name['position_q1'], inplace=True)
    # Asignar la columna 'q3' a 'time'
    df_track_name['time'] = df_track_name['q3']
    # Rellenar los valores faltantes en 'time' con los de 'q2'
    df_track_name['time'].fillna(df_track_name['q2'], inplace=True)
    # Rellenar los valores faltantes en 'time' con los de 'q1'
    df_track_name['time'].fillna(df_track_name['q1'], inplace=True)
    
    # Mapear el driverid a driverref
    df_track_name['driver_name'] = df_track_name['driverid'].map(driverid_to_driverref)
    df_track_name['q3'] = df_track_name['q3'].fillna(0)
    
    columns_to_drop = ['qualifyid', 'raceid', 'constructorid', 'round', 'circuitid', 'car_condition',
                   'driver_condition', 'previous_year_position_q1', 'circuitref_fastlap', 'driver_time_vs_fastlap',
                   'car_time_vs_fastlap', 'previous_driver_time_vs_fastlap', 'previous_car_time_vs_fastlap',
                   'q2disqualified', 'driver_q2vsq1', 'car_q2vsq1', 'previous_driver_q2vsq1', 'previous_car_q2vsq1',
                   'previous_year_position_q2', 'q3disqualified', 'driver_q3vsq2', 'car_q3vsq2', 'previous_driver_q3vsq2',
                   'previous_car_q3vsq2', 'previous_year_position_q3']
    df_track_name.drop(columns=columns_to_drop, inplace=True)
    
    # Convertir las columnas de milisegundos a segundos
    # Convertir las columnas de milisegundos a segundos
    df_track_name['q1'] = df_track_name['q1'] / 1000
    df_track_name['q2'] = df_track_name['q2'] / 1000
    df_track_name['q3'] = df_track_name['q3'] / 1000
    df_track_name['time'] = df_track_name['time'] / 1000

    # Convertir las columnas de segundos a formato de minutos, segundos y milisegundos con tres cifras
    df_track_name['q1'] = pd.to_datetime(df_track_name['q1'], unit='s').dt.strftime('%M:%S.%f')
    df_track_name['q2'] = pd.to_datetime(df_track_name['q2'], unit='s').dt.strftime('%M:%S.%f')
    df_track_name['q3'] = pd.to_datetime(df_track_name['q3'], unit='s').dt.strftime('%M:%S.%f')
    df_track_name['time'] = pd.to_datetime(df_track_name['time'], unit='s').dt.strftime('%M:%S.%f')


 
```
