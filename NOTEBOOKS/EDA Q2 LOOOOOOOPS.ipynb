```python
dfs = {}
missing_count = 0  # Contador para los df_track_name que faltan

for file_name in file_names:
    track_name = file_name.replace(".csv", "")
    df = pd.read_csv(file_name)
    df.sort_values(by=['year', 'driverid'], ascending=[True, True], inplace=True)
    dfs[track_name] = df

diccionario = {"bahrain": "bahrain",
              "albert": "albert_park",
              "losail": "losail",
              "imola": "imola",
              "portimao": "portimao",
              "catalunya": "catalunya",
              "monaco": "monaco",
              "baku": "baku",
              "red": "red_bull_ring",
              "ricard": "ricard",
              "silverstone": "silverstone",
              "hungaroring": "hungaroring",
              "spa": "spa",
              "zandvoort": "zandvoort",
              "monza": "monza",
              "sochi": "sochi",
              "istanbul": "istanbul",
              "americas": "americas",
              "rodriguez": "rodriguez",
              "interlagos": "interlagos",
              "jeddah": "jeddah",
              "yas": "yas_marina",
              "miami": "miami",
              "villeneuve": "villeneuve",
              "marina": "marina_bay",
              "suzuka": "suzuka"}
for track_name, df_track_name in dfs.items():
    variable = track_name.split("_")[1]
    if variable in diccionario.keys():
        variable1 = diccionario[variable]
        
    df3 = df2[df2["circuitref"] == variable1]
    df_track_name = pd.merge(df_track_name, df3[['driverid', 'year', 'q2']], on=['driverid', 'year'], how='left')
    df_track_name['q2disqualified'] = (df_track_name['position_q1'] > 15).astype(int)
    df_track_name["driver_q2vsq1"] = df_track_name["q2"] - df_track_name["q1"]
    df_track_name.loc[df_track_name["q2"] == 0, "driver_q2vsq1"] = 0
    car_q2vsq1 = df_track_name.groupby(['constructorid', 'year', 'circuitref'])['driver_q2vsq1'].mean().reset_index()
    df_track_name = pd.merge(df_track_name, car_q2vsq1, on=['constructorid', 'year', 'circuitref'])
    df_track_name = df_track_name.rename(columns={'driver_q2vsq1_x': 'driver_q2vsq1', 'driver_q2vsq1_y': 'car_q2vsq1'})
    df_track_name['previous_driver_q2vsq1'] = df_track_name.groupby(['circuitref', 'driverid'])['driver_q2vsq1'].shift()
    df_track_name['previous_driver_q2vsq1'] = df_track_name['previous_driver_q2vsq1'].fillna(0)
    df_track_name['previous_car_q2vsq1'] = df_track_name.groupby(['circuitref', 'driverid'])['car_q2vsq1'].shift()
    df_track_name['previous_car_q2vsq1'] = df_track_name['previous_car_q2vsq1'].fillna(0)
    df_track_name['position_q2'] = df_track_name.groupby(['circuitref', 'year'])['q2'].rank()
    df_track_name['previous_year_position_q2'] = df_track_name.groupby(['circuitref', 'driverid'])['position_q2'].shift()
    df_track_name['previous_year_position_q2'] = df_track_name['previous_year_position_q2'].fillna(20)
    column_order = ['qualifyid', 'raceid', 'driverid', 'constructorid', 'position_final',
                        'year', 'round', 'circuitid', 'circuitref', 'car_condition',
                        'driver_condition', 'position_q1', 'previous_year_position_q1',
                        'circuitref_fastlap', 'driver_time_vs_fastlap', 'car_time_vs_fastlap',
                        'previous_driver_time_vs_fastlap', 'previous_car_time_vs_fastlap',
                        'q2disqualified', 'driver_q2vsq1', 'car_q2vsq1',
                        'previous_driver_q2vsq1', 'previous_car_q2vsq1', 'position_q2',
                        'previous_year_position_q2', 'q1', 'q2']
    df_track_name = df_track_name.reindex(columns=column_order)
    df_track_name.to_csv(f'{track_name}_trabajar_posiciones_finales.csv', index=False)
    df_track_name = df_track_name[df_track_name["q2disqualified"] == 0]
    
    
    
    
    max_q2 = df_track_name['q2'].max()
    df_track_name['q2'].replace(0, max_q2, inplace=True)
    
    
    
    
    
    dfs[track_name] = df_track_name

    drop_list = ["q2", "raceid", "position_final", "circuitref", "driver_time_vs_fastlap", "car_time_vs_fastlap",
                     "raceid", "car_condition", "driver_condition", 'driver_q2vsq1', 'car_q2vsq1', "q1",
                     "position_q2"]
    X = df_track_name.drop(drop_list, axis=1)
    y = df_track_name["q2"]
    X_2024 = X[X['year'] == 2024]
    y_2024 = y[X['year'] == 2024]
    X_train, X_test, y_train, y_test = train_test_split(X[X['year'] != 2024], y[X['year'] != 2024],
                                                           test_size=0.02, random_state=0)

    xgbr = xgb.XGBRFRegressor()
    xgbr.fit(X_train, y_train)
    predictions_xgb = xgbr.predict(X_test)
    predictions_xgb_2024 = xgbr.predict(X_2024)
    r2_3_6 = r2_score(y_test, predictions_xgb)
    RMSE_3_6 = mean_squared_error(y_test, predictions_xgb, squared=False)
    MSE_3_6 = mean_squared_error(y_test, predictions_xgb)
    MAE_3_6 = mean_absolute_error(y_test, predictions_xgb)
    series_pred_2024 = pd.Series(predictions_xgb_2024, name='q2')
    duplicates = series_pred_2024[series_pred_2024.duplicated(keep=False)]
    duplicate_count = 0
    for index in duplicates.index:
        duplicate_count += 1
        value_to_add = duplicate_count - 1
        series_pred_2024[index] += value_to_add
    df_track_name_2024 = pd.concat([X_2024.reset_index(drop=True)], axis=1)
    df_track_name_2024['q2'] = series_pred_2024
    
    
    df_track_name_2024['position_q2'] = df_track_name_2024['q2'].rank(method='min')
    
    
    
    
    df_track_name_2024.to_csv(f'{track_name}_q2_predicted.csv', index=False)
    if df_track_name.empty:
        missing_count += 1
print(f"Total de df_track_name faltantes: {missing_count}")
```
