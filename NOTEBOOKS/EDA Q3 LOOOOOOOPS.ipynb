```python
diccionario = {
    "bahrain": "bahrain",
    "albert": "albert_park",
    "losail": "losail",
    "imola": "imola",
    "portimao": "portimao",
    "catalunya": "catalunya",
    "monaco": "monaco",
    "baku": "baku",
    "red": "red_bull_ring",
    "ricard": "ricard",
    "silverstone": "silverstone",
    "hungaroring": "hungaroring",
    "spa": "spa",
    "zandvoort": "zandvoort",
    "monza": "monza",
    "sochi": "sochi",
    "istanbul": "istanbul",
    "americas": "americas",
    "rodriguez": "rodriguez",
    "interlagos": "interlagos",
    "jeddah": "jeddah",
    "yas": "yas_marina",
    "miami": "miami",
    "villeneuve": "villeneuve",
    "marina": "marina_bay",
    "suzuka": "suzuka"}
# Crear una copia del diccionario dfs antes de iterar sobre √©l
dfs_copy = dfs.copy()
missing_count = 0
# Iterar sobre la copia del diccionario
for key, df in dfs_copy.items():
    variable = key.split("_")[1]
    if variable in diccionario:
        variable1 = diccionario[variable]
        df3 = df2[df2["circuitref"] == variable1]
        dfs[key] = pd.merge(dfs[key], df3[['year', 'driverid', 'q3']], on=['year', 'driverid'], how='left')
        dfs[key]['q3disqualified'] = (dfs[key]['position_q2'] > 10).astype(int)
        dfs[key]["driver_q3vsq2"] = dfs[key]["q3"] - dfs[key]["q2"]
        dfs[key].loc[dfs[key]["q3"] == 0, "driver_q3vsq2"] = 0 
        car_q3vsq2 = dfs[key].groupby(['constructorid', 'year', 'circuitref'])['driver_q3vsq2'].mean().reset_index()
        dfs[key] = pd.merge(dfs[key], car_q3vsq2, on=['constructorid', 'year', 'circuitref'], suffixes=('', '_mean'))
        dfs[key].rename(columns={'driver_q3vsq2_x': 'driver_q3vsq2', 'driver_q3vsq2_mean': 'car_q3vsq2'}, inplace=True)
        dfs[key]['previous_driver_q3vsq2'] = dfs[key].groupby(['circuitref', 'driverid'])['driver_q3vsq2'].shift().fillna(0)
        dfs[key]['previous_car_q3vsq2'] = dfs[key].groupby(['circuitref', 'driverid'])['car_q3vsq2'].shift().fillna(0)
        dfs[key]['position_q3'] = dfs[key].groupby(['circuitref', 'year'])['q3'].rank() 
        dfs[key]['previous_year_position_q3'] = dfs[key].groupby(['circuitref', 'driverid'])['position_q3'].shift().fillna(20)
        column_order = ['qualifyid', 'raceid', 'driverid', 'constructorid', 'position_final',
                        'year', 'round', 'circuitid', 'circuitref', 'car_condition',
                        'driver_condition', 'position_q1', 'previous_year_position_q1',
                        'circuitref_fastlap', 'driver_time_vs_fastlap', 'car_time_vs_fastlap',
                        'previous_driver_time_vs_fastlap', 'previous_car_time_vs_fastlap',
                        'q2disqualified', 'driver_q2vsq1', 'car_q2vsq1',
                        'previous_driver_q2vsq1', 'previous_car_q2vsq1', 'position_q2',
                        'previous_year_position_q2', 'q3disqualified',
                        'driver_q3vsq2', 'car_q3vsq2', 'previous_driver_q3vsq2',
                        'previous_car_q3vsq2', 'position_q3', 'previous_year_position_q3', 'q1', 'q2', 'q3']
        dfs[key] = dfs[key].reindex(columns=column_order)
        dfs[key].to_csv(f'{key}_penultimo.csv', index=False)
        dfs[key] = dfs[key][dfs[key]["q3disqualified"] == 0]
        
        
        
        max_q3 = dfs[key]['q3'].max()
        dfs[key]['q3'].replace(0, max_q3, inplace=True)       
        
        
        
        
        drop_list = ["position_final", "circuitref", "driver_time_vs_fastlap",
                    "car_time_vs_fastlap", "raceid", "car_condition", 'position_q3',
                    "driver_condition", 'driver_q2vsq1', 'car_q2vsq1', 'q1', 'q2', 'q3',
                    "car_q3vsq2", "driver_q3vsq2"]
        X = dfs[key].drop(drop_list, axis=1)
        y = dfs[key]["q3"]
        X_2024 = X[X['year'] == 2024]
        X_train, X_test, y_train, y_test = train_test_split(X[X['year'] != 2024], y[X['year'] != 2024],
                                                            test_size=0.02, random_state=0)

        xgbr = xgb.XGBRFRegressor()
        xgbr.fit(X_train, y_train)
        predictions_xgb_2024 = xgbr.predict(X_2024)
        series_pred_2024 = pd.Series(predictions_xgb_2024, name='q3')
        duplicates = series_pred_2024[series_pred_2024.duplicated(keep=False)]
        duplicate_count = 0
        for index in duplicates.index:
            duplicate_count += 1
            value_to_add = duplicate_count - 1
            series_pred_2024[index] += value_to_add 
        dfs[f"{key}_2024"] = pd.concat([X_2024.reset_index(drop=True), series_pred_2024], axis=1)
        
        
        
        dfs[f"{key}_2024"]['position_q3'] = dfs[f"{key}_2024"]['q3'].rank(method='min')
        
        
        
        dfs[f"{key}_2024"].to_csv(f'{key}_q3_predicted_2024.csv', index=False)
        if dfs[key].empty:
            missing_count += 1
print(f"Total de df_track_name faltantes: {missing_count}")
```
