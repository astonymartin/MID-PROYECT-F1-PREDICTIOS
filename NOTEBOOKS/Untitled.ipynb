```python

import matplotlib.pyplot as plt

# Definir el diccionario con los datos
diccionario2 = {"valores creados": 1580*30,
              "valores descargados": 1110*9}

# Convertir el diccionario en listas separadas para las etiquetas (x) y los valores (y)
etiquetas = list(diccionario2.keys())
valores = list(diccionario2.values())

# Crear el gr√°fico de l√≠neas
plt.plot(etiquetas, valores, marker='o', linestyle='-')

# A√±adir etiquetas al eje x y al t√≠tulo
plt.xticks(rotation=45, ha='right')
plt.xlabel('Variables')
plt.ylabel('Valores')
plt.title('Gr√°fico de L√≠neas de Variables')

# Mostrar el gr√°fico de l√≠neas
plt.grid(True)
plt.show()
```
